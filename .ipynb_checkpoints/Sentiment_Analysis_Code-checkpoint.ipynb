### Sentiment Analysis: An Overview

**1. What is Sentiment Analysis?**
Sentiment analysis, also known as opinion mining, is a Natural Language Processing (NLP) technique used to determine the sentiment expressed in a piece of text. It involves identifying the emotions, opinions, or attitudes conveyed, classifying them as positive, negative, neutral, or sometimes more nuanced categories like very positive, very negative, etc. Sentiment analysis is applied to text data from various sources, including social media, reviews, articles, and more.

**2. Key Aspects of Sentiment Analysis:**
   
   - **Sentiment Classification:** The primary goal is to classify a given text into predefined sentiment categories. The simplest model might classify texts as positive, negative, or neutral, while more advanced models can include finer distinctions.

   - **Polarity Detection:** This aspect focuses on determining the positivity or negativity of a sentiment. For example, in product reviews, polarity detection can identify whether a review is favorable or unfavorable.

   - **Emotion Detection:** Beyond just polarity, sentiment analysis can be used to detect specific emotions expressed in the text, such as joy, anger, sadness, fear, etc. This requires a more detailed understanding of the nuances of language.

   - **Aspect-based Sentiment Analysis:** This involves analyzing sentiment towards specific aspects or features of a product or service. For instance, in a review of a smartphone, the sentiment towards the camera, battery life, design, and price might all be separately analyzed.

   - **Subjectivity/Objectivity Analysis:** This determines whether a statement is subjective (based on personal opinions) or objective (based on factual information).

**3. Techniques and Methods:**

   - **Lexicon-based Approaches:** This method relies on a predefined list of words with associated sentiment values. The overall sentiment of a text is determined based on the sentiment values of the words it contains.

   - **Machine Learning Approaches:** These methods involve training models on labeled datasets where the sentiment is already known. Common algorithms include Naive Bayes, Support Vector Machines (SVM), and more recently, deep learning techniques like Recurrent Neural Networks (RNNs) and transformers.

   - **Hybrid Approaches:** These combine lexicon-based and machine learning approaches to leverage the strengths of both.

**4. Applications of Sentiment Analysis:**

   - **Customer Feedback and Product Reviews:** Companies use sentiment analysis to understand customer opinions about their products or services. This helps in product improvement, marketing strategies, and customer support.

   - **Social Media Monitoring:** Sentiment analysis is used to gauge public opinion on various topics, from political events to brand perception. It helps organizations monitor and manage their reputation.

   - **Market Research:** Businesses use sentiment analysis to analyze market trends and consumer behavior, aiding in decision-making and strategic planning.

   - **Financial Markets:** Sentiment analysis of news articles, social media, and financial reports can provide insights into market sentiment, helping investors make informed decisions.

   - **Healthcare:** In healthcare, sentiment analysis can be used to analyze patient feedback, understand patient satisfaction, and even monitor public health concerns through social media.

   - **Political Analysis:** Sentiment analysis can be applied to political discourse to understand public opinion on policies, candidates, or events, providing valuable insights for political strategists and analysts.

   - **Content Moderation:** It can be used to detect harmful or inappropriate content, such as hate speech or harassment, on social media platforms.

**5. Challenges in Sentiment Analysis:**

   - **Sarcasm and Irony:** Detecting sarcasm and irony is challenging because the literal meaning of words often differs from the intended sentiment.

   - **Context and Ambiguity:** Words and phrases can have different meanings depending on the context, making accurate sentiment detection difficult.

   - **Language and Cultural Differences:** Sentiment expressions can vary widely across different languages and cultures, requiring tailored approaches.

   - **Domain-Specific Sentiment:** Words may carry different sentiments in different domains (e.g., "cold" in a weather context vs. a human interaction context), necessitating domain-specific models.

**6. Future Directions:**

   - **Enhanced Emotion Detection:** Improving the granularity of emotion detection to capture a wider range of human emotions.

   - **Multimodal Sentiment Analysis:** Combining textual data with other modalities like images, audio, or video to provide a more comprehensive sentiment analysis.

   - **Explainability:** Developing methods to make sentiment analysis models more interpretable and explainable, especially in critical applications like healthcare and finance.

   - **Real-time Sentiment Analysis:** Enhancing the speed and accuracy of real-time sentiment analysis for applications like social media monitoring and customer service.

Sentiment analysis is a powerful tool that leverages advancements in NLP and machine learning to provide valuable insights across various domains. Its applications continue to expand as technology advances and the importance of understanding public sentiment grows.

**Deep learning classifiers are algorithms or models used in deep learning to classify data into different categories or classes.**

    These models are typically built using neural networks, particularly deep neural networks with multiple layers. Here are some common types of deep learning classifiers:

 1. **Multilayer Perceptron (MLP)**
    An MLP is a simple type of neural network that consists of an input layer, one or more hidden layers, and an output layer. It is fully connected, meaning each neuron in one layer is connected to every neuron in the next layer. MLPs are capable of learning complex patterns but are generally less powerful than more advanced architectures.
    


 2. **Convolutional Neural Networks (CNNs)**
    CNNs are specifically designed to process data with a grid-like topology, such as images. They use convolutional layers to extract features from the input data. Key components include:

       - **Convolutional Layers:** Apply convolution operations to detect features.
       - **Pooling Layers:** Downsample the data to reduce dimensionality and computational load.
       - **Fully Connected Layers:** Serve as the classifier by mapping the extracted features to the output classes.

    CNNs are widely used for image and video recognition, object detection, and other tasks involving spatial data.
    
    

 3. **Recurrent Neural Networks (RNNs)**
    RNNs are designed for sequential data and are commonly used in tasks like natural language processing, time series prediction, and speech recognition. They have loops that allow information to persist, making them capable of learning temporal dependencies. Key variants include:

       - **Long Short-Term Memory (LSTM):** A type of RNN that can learn long-term dependencies using gates to control the flow of information.
       - **Gated Recurrent Units (GRU):** A simplified version of LSTM that is computationally efficient and performs well on many tasks.
       
       

 4. **Transformer Networks**
    Transformers are a type of neural network architecture introduced for natural language processing tasks. They use self-attention mechanisms to weigh the importance of different words in a sentence or parts of an input sequence. Transformers are the backbone of many state-of-the-art NLP models, such as BERT and GPT.



 5. **Autoencoders**
    Autoencoders are neural networks used for unsupervised learning. They consist of an encoder and a decoder. The encoder compresses the input into a lower-dimensional representation, and the decoder reconstructs the input from this representation. While primarily used for feature extraction and dimensionality reduction, autoencoders can also serve as classifiers when paired with other techniques.



 6. **Generative Adversarial Networks (GANs)**
    GANs consist of two networks: a generator and a discriminator. The generator creates fake data samples, while the discriminator attempts to distinguish between real and fake samples. Although GANs are primarily used for generating data, the discriminator can be adapted for classification tasks.



 7. **Capsule Networks (CapsNets)**
    Capsule Networks aim to address some limitations of CNNs, particularly in understanding spatial hierarchies in data. They use capsules—groups of neurons that output vectors instead of scalars—to capture and represent the spatial relationships between features. CapsNets have shown promise in tasks requiring a more nuanced understanding of spatial relationships.



 8. **Deep Belief Networks (DBNs)**
    DBNs are generative models composed of multiple layers of hidden units, where each pair of layers forms a Restricted Boltzmann Machine (RBM). They are used for feature learning and classification, though they have been largely superseded by other deep learning methods.



 9. **Deep Convolutional Generative Adversarial Networks (DCGANs)**
    A variant of GANs that uses convolutional layers in both the generator and discriminator. While mainly used for generating realistic images, the discriminator can be employed as a classifier.



 10. **Attention Mechanisms**
        Attention mechanisms are often used in conjunction with other architectures, like RNNs or Transformers. They allow the model to focus on specific parts of the input when making decisions, improving performance on tasks like language translation and text classification.



 11. **Self-Supervised Learning Models**
     These models, like BERT and GPT, are pre-trained on large datasets in a self-supervised manner and then fine-tuned for specific classification tasks. They have shown state-of-the-art performance in various NLP applications.

    Deep learning classifiers have revolutionized fields like computer vision, natural language processing, speech recognition, and more, due to their ability to automatically learn and extract complex patterns from data.

#### As we can see that, we have 6 labels or targets in the dataset. We can make a multi-class classifier for Sentiment Analysis. But, for the sake of the simplicity, we will merge these labels into two classes, i.e. Positive or Negative sentiment.

    1. Positive Sentiment - "joy","love","surprise"
    2. Negative Sentiment - "anger","sadness","fear"

#### Now, we will create a custom encoder to convert categorical target labels to numerical form, i.e. (0 and 1)

```python
def custom_encoder(df):
    df.replace(to_replace ="surprise", value =1, inplace=True)
    df.replace(to_replace ="love", value =1, inplace=True)
    df.replace(to_replace ="joy", value =1, inplace=True)
    df.replace(to_replace ="fear", value =0, inplace=True)
    df.replace(to_replace ="anger", value =0, inplace=True)
    df.replace(to_replace ="sadness", value =0, inplace=True)
```

#### Now, we will perform some pre-processing on the data before converting it into vectors and passing it to the machine learning model.

```python
#object of WordNetLemmatizer
lm = WordNetLemmatizer()
```

#### We will create a function for pre-processing of data
1. First, we will iterate through each record and using regualr expression, we will get rid of any characters apart from alphabets.
2. Then, we will convert the string to lowercase as, the word "Good" is different from the word "good". This will cause an issue, when we will create vectors of these words, as two different vectors will be created for the same word which we don't want to.
3. Then we will check for stopwords in the data and get rid of them. Stopwords are commonly used words in a sentence such as "the","an","to" etc. which do not add much value.
4. Then, we will perform lemmatization on each word,i.e. change the different forms of word into a single item called as lemma. A lemma is a base form of a word. For example, run, running and runs are all forms of same lexeme where run is the lemma. Hence, we are converting all occurrences of same lexeme to it's respective lemma.
5. And, then return a corpus of processed data

```python
def text_transformation(df_col):
    corpus = []
    for item in df_col:
        new_item = re.sub('[^a-zA-Z]',' ',str(item))
        new_item = new_item.lower()
        new_item = new_item.split()
        new_item = [lm.lemmatize(word) for word in new_item if word not in set(stopwords.words('english'))]
        corpus.append(' '.join(str(x) for x in new_item))
    return corpus
```

```python
corpus = text_transformation(df['text'])
```

#### Now, we will convert the text data into vectors, by fitting and transforming the corpus.
#### We will take ngram_range as (1,2) which signifies a bigram. 
#### Ngram is a sequence of 'n' of words in a row or sentence. 'ngram_range' is parameter which we use to give importance to combination of words as well, such as, "social media" has different meaning then separate words such as "social" and "media".
#### We can use experiment with the value of this parameter and select the option which gives better results.


```python
cv = CountVectorizer(ngram_range=(1,2))
traindata = cv.fit_transform(corpus)
```

```python
X = traindata
y = df.label
```

```python
print(traindata)
```

#### In this project, I'm going to use Random Forest Classifier, and we will tune the hyperparameters using GridSearchCV.

#### We will create a dictionary, "parameters" which will contain the values of different hyperparameters, which we will pass to GridSearchCV to train our random forest classifier model using all possible combinations of the parameters.

```python
parameters = {'max_features': ('auto','sqrt'),
             'n_estimators': [500, 1000],
             'max_depth': [10, None],
             'min_samples_split': [5],
             'min_samples_leaf': [1],
             'bootstrap': [True]}
```

#### Now, our GridSearchCV() will take the following parameters,
    1. Estimator or model - RandomForestClassifier in our case
    2. parameters - dictionary of hyperparameter names and their values
    3. cv - signifies cross validation folds
    4. return_train_score - returns the training scores of the various maodels
    5. n_jobs - no. of jobs to run parallely ("-1" signifies that all CPU cores will be used which reduces the training time drastically)

#### Now, we will fit the data into grid search 

        - the below code snippet will take round about 6 hours

```python
grid_search = GridSearchCV(RandomForestClassifier(),parameters,cv=5,return_train_score=True,n_jobs=-1)
grid_search.fit(X,y)
grid_search.best_params_
```

#### Displays models and their respective parameters, mean test score and rank. 

```python
for i in range(6):
    print('Parameters: ',grid_search.cv_results_['params'][i])
    print('Mean Test Score: ',grid_search.cv_results_['mean_test_score'][i])
    print('Rank: ',grid_search.cv_results_['rank_test_score'][i])
```

#### Now, we will choose the best parameters obtained from GridSearchCV and create a final random forest classifier model

```python
rfc = RandomForestClassifier(max_features=grid_search.best_params_['max_features'],
                                      max_depth=grid_search.best_params_['max_depth'],
                                      n_estimators=grid_search.best_params_['n_estimators'],
                                      min_samples_split=grid_search.best_params_['min_samples_split'],
                                      min_samples_leaf=grid_search.best_params_['min_samples_leaf'],
                                      bootstrap=grid_search.best_params_['bootstrap'])
```

#### Now, we will train our new model.

```python
rfc.fit(X,y)
```

## Test Data Transformation
#### Now, we will read the test data and perform the same transformations we did on training data and finally evaluate the model on it's predictions.

```python
test_df = pd.read_csv('test.txt',delimiter=';',names=['text','label'])
```

```python
test_df.head()
```

```python
X_test,y_test = test_df.text,test_df.label
#encode the labels into two classes , 0 and 1
test_df = custom_encoder(y_test)
#pre-processing of text
test_corpus = text_transformation(X_test)
#convert text data into vectors
testdata = cv.transform(test_corpus)
#predict the target
predictions = rfc.predict(testdata)
```

### Model Evaluation Metrics

```python
rcParams['figure.figsize'] = 10,5
plot_confusion_matrix(y_test,predictions)
acc_score = accuracy_score(y_test,predictions)
pre_score = precision_score(y_test,predictions)
rec_score = recall_score(y_test,predictions)
print('Accuracy_score: ',acc_score)
print('Precision_score: ',pre_score)
print('Recall_score: ',rec_score)
print('-------------------------------------------------------------------')
cr = classification_report(y_test,predictions)
print(cr)
```

```python
predictions_probability = rfc.predict_proba(testdata)
```

```python
fpr,tpr,thresholds = roc_curve(y_test,predictions_probability[:,1])
plt.plot(fpr,tpr)
plt.plot([0,1])
plt.title('ROC Curve')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.show()
```

## Predict for Custom Input

```python
def expression_check(prediction_input):
    if prediction_input == 0:
        print("Input statement has Negative Sentiment.")
    elif prediction_input == 1:
        print("Input statement has Positive Sentiment.")
    else:
        print("Invalid Statement.")
```

```python
def sentiment_predictor(input):
    input = text_transformation(input)
    transformed_input = cv.transform(input)
    prediction = rfc.predict(transformed_input)
    expression_check(prediction)
```
